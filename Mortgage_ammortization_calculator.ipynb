## 30-year Mortgage Fixed at 4 percent

```python
import numpy_financial as npf
import pandas as pd
import numpy as np
import math
```

```python
#Find out functions you can use with numpy_financials using list comprehension
[function for function in dir(npf) if not function.startswith ('_')]
```

```python
#Defining variables for 30 year mortgage at 4% fixed interest rate
p= 1000000 
r = 0.04 
t= 30 

print('{:^18}{:^18}{:^20}{:^18}{:^18}{:^18}'.format('Month Number','New Principal', 'Interest Applied', 'Principle Paydown', 'Fixed Payment',  'Unpaid Principal Balance'))

monthNo = t*12

#Fixed installments that needs to be made every month
fixedPayment=(p*(r/12)*(math.pow(1+r/12,12*t)))/(math.pow(1+r/12,12*t)-1)


newPrincipal = p
unpaidPrincipalBalance = p
for i in range(1,monthNo+1):
    
    interestApplied = r/12*newPrincipal
    principalToPay = fixedPayment-interestApplied
    unpaidPrincipalBalance=newPrincipal+interestApplied-fixedPayment
    a = ('{:^18d}{:^18.2f}{:^20.2f}{:^18.2f}{:^18.2f}{:^18.2f}'.format(i,newPrincipal,interestApplied,principalToPay,fixedPayment,unpaidPrincipalBalance,))
    print(a)
        
    newPrincipal=unpaidPrincipalBalance
   
```

## Calculate it yourself (Input Principal, Rate & Term to Calculate) 

```python
# Input Variables needed to be entered by user
p= float(input("Principal $: ")) #Principal = 1000000
r = float(input("Rate: ")) #Rate = 0.04 
t= int (input("Term in number of years: ")) #Term = 30

#print('{:^18}{:^18}{:^20}{:^18}{:^18}{:^18}'.format('Month Number','New Principal', 'Interest Applied', 'Principle Paydown', 'Fixed Payment',  'Unpaid Principal Balance'))
monthNo = t*12
fixedPayment=(p*(r/12)*(math.pow(1+r/12,12*t)))/(math.pow(1+r/12,12*t)-1)



df6=[]

newPrincipal = p
unpaidPrincipalBalance = p
for i in range(1,monthNo+1):
    
    interestApplied = r/12*newPrincipal
    principalPaydown = fixedPayment-interestApplied
    unpaidPrincipalBalance=newPrincipal+interestApplied-fixedPayment
    #a = ('{:^18d}{:^18.2f}{:^20.2f}{:^18.2f}{:^18.2f}{:^18.2f}'.format(i,newPrincipal,interestApplied,principalToPay,fixedPayment,unpaidPrincipalBalance,))
    #print(a)
    
    df5 = [i,newPrincipal,interestApplied,principalPaydown,fixedPayment,unpaidPrincipalBalance]
    for i in df5:
        df6.append(df5)
    
                       
    
    newPrincipal=unpaidPrincipalBalance
df7 = pd.DataFrame(df6, columns=['monthNumber','newPrincipal','interestApplied','principalPaydown','fixedPayment','unpaidPrincipalBalance'])

df7.set_index("monthNumber")
df8 = df7.groupby("monthNumber")   
```

```python
amortization_30fixed = df8.first()
```
